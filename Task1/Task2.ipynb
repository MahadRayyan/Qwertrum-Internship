```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Ruhan\Desktop\task1\Messy_Employee_dataset.csv", encoding='cp1252')
```

## Problems Found Before Cleaning

| # | Column | Issue | Type |

| 1 | Age | 211 missing values (20% of data) | Missing Values |
| 2 | Salary | 24 missing values | Missing Values |
| 3 | Phone | All values are negative numbers (e.g. -1651623197) | Wrong Data |
| 4 | Join_Date | Stored as string, not datetime | Wrong Data Type |
| 5 | Department_Region | Two columns merged into one (e.g. "DevOps-California") | Wrong Format |

```python
print("=== MISSING VALUES ===")
print(df.isnull().sum())

print("\n=== DUPLICATE ROWS ===")
print(f"Duplicates: {df.duplicated().sum()}")

print("\n=== WRONG DATA TYPES ===")
print(f"Join_Date type: {df['Join_Date'].dtype}  <-- should be datetime")
print(f"Phone sample: {df['Phone'].head(3).tolist()}  <-- negative values!")

print("\n=== DEPARTMENT_REGION MERGED COLUMN ===")
print(df['Department_Region'].unique()[:5])
```

```python
print(f"Rows before: {len(df)}")

# Fill Age with median (best for skewed numeric data)
df['Age'] = df['Age'].fillna(df['Age'].median())
print(f"Age missing after fill: {df['Age'].isnull().sum()}")

# Fill Salary with median (avoids outlier influence)
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
print(f"Salary missing after fill: {df['Salary'].isnull().sum()}")

print("\nMissing values after fix:")
print(df.isnull().sum())
```

```python
print(f"Rows before removing duplicates: {len(df)}")
df = df.drop_duplicates()
print(f"Rows after removing duplicates: {len(df)}")
```

```python
# Fix Join_Date: string â†’ datetime
df['Join_Date'] = pd.to_datetime(df['Join_Date'])
print(f"Join_Date type now: {df['Join_Date'].dtype}")

# Fix Phone: all negative â†’ make positive
df['Phone'] = df['Phone'].abs()
print(f"Phone sample after fix: {df['Phone'].head(3).tolist()}")

# Fix Age: float â†’ int
df['Age'] = df['Age'].astype(int)
print(f"Age type now: {df['Age'].dtype}")
```

```python
# Split "DevOps-California" into Department and Region
df[['Department', 'Region']] = df['Department_Region'].str.split('-', expand=True)
df = df.drop(columns=['Department_Region'])

print("New columns created:")
print(df[['Department', 'Region']].head(5))
```

```python
#Cleaned file
df.to_csv(r"C:\Users\Ruhan\Desktop\task1\Cleaned_Employee_dataset.csv", index=False)
print("Cleaned file saved successfully!")
```
